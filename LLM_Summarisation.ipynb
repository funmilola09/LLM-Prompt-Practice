```python
import os
import json
from openai import OpenAI
from google.colab import userdata

client = OpenAI(
    base_url="https://openrouter.ai/api/v1",
    api_key=userdata.get('open_router'))
```

```python
articles_folder = "/content/sample_data/article"
```

```python
all_summaries = {} # dictionary to hold the summaries
for filename in sorted(os.listdir(articles_folder)): #for loop to loop through the folder that has each files
    if filename.endswith(".txt"): #confirm the files are text files before summarising
        file_path = os.path.join(articles_folder, filename)
        with open(file_path, "r", encoding="latin-1") as f:
             article = f.read() #read the article

        response = client.chat.completions.create( #specify the model to be used for the summarisation
           model="openrouter/free",
           messages=[ #model chat instruction
               {
                    "role": "system",
                    "content": "Summarize articles in 5 bullet points."
               },
               {
                    "role": "user",
                    "content": article
               }
                 ]
        )
        #response


       summary = response.choices[0].message.content
#summary = response.output_text

        all_summaries[filename] = summary

        print(f"\n{'='*60}")
        print(f"SUMMARY FOR: {filename}")
        print(summary)

with open("summaries.json", "w", encoding="utf-8") as f:
    json.dump(all_summaries, f, indent=4, ensure_ascii=False)

print(f"\n Saved{len(all_summaries)} summaries to summaries.json")
```
