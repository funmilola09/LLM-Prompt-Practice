```python
import os
import json
from openai import OpenAI
from google.colab import userdata

client = OpenAI(
    base_url="https://openrouter.ai/api/v1",
    api_key=userdata.get('open_router'))
```

```python
articles = "/content/sample_data/article 1.txt"
```

#Text Summarisation

```python
import time

models = [
    "qwen/qwen3-32b",
    "deepseek/deepseek-r1",
    "meta-llama/llama-3.3-70b-instruct",
    "google/gemma-3-27b-it",
    "mistralai/ministral-3b-2512"
]

all_summaries = {}
all_results = {}

with open(articles, "r", encoding="latin-1") as f:
    article = f.read()

for model in models:
    print(f"Summarising with {model}...")

    start = time.time()

    response = client.chat.completions.create(
        model=model,
        messages=[
            {
                "role": "system",
                "content": "Summarize the following article in exactly five concise bullet points."
            },
            {
                "role": "user",
                "content": article
            }
        ]
    )

    elapsed = time.time() - start

    summary = response.choices[0].message.content

    all_summaries[model] = {
        "summary": summary,
        "time": elapsed
    }

    print(f"{model} completed in {elapsed:.2f} seconds.\n")
```

#Text summarisation Evaluation
this is based on time in seconds, latency, token usage

```python
for model, result in all_summaries.items():
    print("*" * 80)
    print(model)
    print(f"Time: {result['time']:.2f}s")
    print(result["summary"])
    print()
```

```python
all_results[model] = {
    "summary": summary,
    "latency": elapsed,
    "input_tokens": response.usage.prompt_tokens,
    "output_tokens": response.usage.completion_tokens,
    "total_tokens": response.usage.total_tokens,
}
```

```python
for model, result in all_results.items():
    print("=" * 80)
    print(f"Model: {model}")
    print(f"Latency: {result['latency']:.2f} seconds")
    print(f"Input Tokens: {result['input_tokens']}")
    print(f"Output Tokens: {result['output_tokens']}")
    print(f"Total Tokens: {result['total_tokens']}")
    print("\nSummary:")
    print(result["summary"])
    print()
```

#Data Generation using function call

```python
import time

def evaluate_models(models, system_prompt, user_prompt):
    results = {}

    for model in models:

        start = time.time()

        response = client.chat.completions.create(
            model=model,
            messages=[
                {
                    "role": "system",
                    "content": system_prompt
                },
                {
                    "role": "user",
                    "content": user_prompt
                }
            ]
        )

        elapsed = time.time() - start

        results[model] = {
            "response": response.choices[0].message.content,
            "latency": elapsed,
            "prompt_tokens": response.usage.prompt_tokens,
            "completion_tokens": response.usage.completion_tokens,
            "total_tokens": response.usage.total_tokens
        }

    return results
```

```python
def print_results(results):

    for model, result in results.items():

        print("=" * 80)
        print(f"Model: {model}")
        print(f"Latency: {result['latency']:.2f} seconds")
        print(f"Prompt Tokens: {result['prompt_tokens']}")
        print(f"Completion Tokens: {result['completion_tokens']}")
        print("-" * 80)
        print(result["response"])
        print()
```

```python
system_prompt = (
    "You are an expert synthetic data generator."
)

prompt = """
Generate a synthetic dataset of 10 employees.

Output only valid CSV.

Columns:
EmployeeID, Age, Department, Salary
"""

generation_results = evaluate_models(
    models,
    system_prompt,
    prompt
)

print_results(generation_results)
```

#Code Generation

```python
system_prompt = (
    """ You are an expert Python programmer.
    Return only Python code.

Do not include explanations, markdown, or surrounding text."""
)

prompt = """
Write a complete Python program that classifies SMS messages as Spam or Ham.

Requirements:
- Use Python.
- Use scikit-learn.
- Use CountVectorizer.
- Use MultinomialNB.
- Include data loading.
- Include model training.
- Include prediction.
- Include comments explaining the code.
- Ensure the code is executable.
"""

data_generator = evaluate_models(
    models,
    system_prompt,
    prompt
)

print_results(data_generator)
```
